```python
import pandas as pd 
import numpy as np
from datetime import datetime
```

```python
chicago = pd.read_csv('bikeshare/chicago.csv')
```

```python
chicago.info()
```

```python
chicago.head()
```

```python
chicago['Start Time']
```

```python
datetime_chicago_list = []
for i in chicago['Start Time']:
    dt_chicago = datetime.strptime(str(i), "%Y-%m-%d %H:%M:%S")
    datetime_chicago_list.append(dt_chicago)
```

```python
datetime_chicago_list
```

```python
chicago_dates_df = pd.DataFrame({'chicago_dates':datetime_chicago_list})
```

```python
chicago_dates_df
```

```python
from statistics import mode

month_ch =[]
i = 0
for i in range(300000):
    m_ch = datetime_chicago_list[i].month
    month_ch.append(int(m_ch))
    i+=1
```

# Most Common Month in Chicago

```python
mode(month_ch)
```

```python

days_ch =[]
i = 0

for i in range(300000):
    d_ch = datetime_chicago_list[i].weekday()
    days_ch.append(str(d_ch))
    i+=1
```

```python
days_ch
```

# Most Common Day of week in Chicago

```python
mode(days_ch)
```

```python
hour_ch =[]
i = 0

for i in range(300000):
    h_ch = datetime_chicago_list[i].hour
    hour_ch.append(str(h_ch))
    i+=1
```

```python
hour_ch
```

# Most Common hour in Chicago

```python
mode(hour_ch)
```

# Most Common 2 Stations in Chicago

```python
freq_ch_Stations =[]
i = 0

for i in range(300000):   
    freq_ch_Stations.append(str(str(chicago['End Station'][i])+str(chicago['Start Station'][i])))
    i+=1
```

```python
mode(freq_ch_Stations)
```

# Total Travel Durations in Chicago

```python
total_travel_time = []
    
for i in range(300000):   
    total_travel_time.append(int(chicago['Trip Duration'][i])/60)
    i+=1
```

```python
total_travel_time
```

```python
travel_time_df = pd.DataFrame({'chicago_trav_time':total_travel_time})
```

# Average Travel Durations in Chicago

```python
travel_time_df.mean()
```
