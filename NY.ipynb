```python
import pandas as pd 
import numpy as np
from datetime import datetime
```

```python
NY = pd.read_csv('bikeshare/new_york_city.csv')
```

```python
NY.info()
```

```python
NY.head()
```

```python
NY['Start Time']
```

```python
datetime_NY_list = []
for i in NY['Start Time']:
    dt_NY = datetime.strptime(str(i), "%Y-%m-%d %H:%M:%S")
    datetime_NY_list.append(dt_NY)
```

```python
datetime_NY_list
```

```python
NY_dates_df = pd.DataFrame({'NY_dates':datetime_NY_list})
```

```python
NY_dates_df
```

```python
from statistics import mode

month_NY =[]
i = 0
for i in range(300000):
    m_ch = datetime_NY_list[i].month
    month_NY.append(int(m_ch))
    i+=1
```

# Most Common Month in NY

```python
mode(month_NY)
```

```python

days_NY =[]
i = 0

for i in range(300000):
    d_NY = datetime_NY_list[i].weekday()
    days_NY.append(str(d_NY))
    i+=1
```

```python
days_NY
```

# Most Common Day of week in NY

```python
mode(days_NY)
```

```python
hour_NY =[]
i = 0

for i in range(300000):
    h_NY = datetime_NY_list[i].hour
    hour_NY.append(str(h_NY))
    i+=1
```

```python
hour_NY
```

# Most Common hour in NY

```python
mode(hour_NY)
```

# Total Travel Durations in NY

```python
total_travel_time = []
    
for i in range(300000):   
    total_travel_time.append(int(NY['Trip Duration'][i])/60)
    i+=1
```

```python
total_travel_time
```

```python
travel_time_df = pd.DataFrame({'NY_trav_time':total_travel_time})
```

# Average Travel Durations in NY
